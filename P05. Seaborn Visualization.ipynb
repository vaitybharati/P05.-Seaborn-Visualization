# Seaborn Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from vega_datasets import data
```

```python
barley=data('barley')
```

```python
barley
```

## Univariate → “one variable” data visualization

### Strip Plot

```python
sns.stripplot(y='yield',data=barley,jitter=True)
```

```python
sns.stripplot(y='yield',data=barley,jitter=False)
```

```python
sns.stripplot(y='yield',data=barley,jitter=True)
plt.ylabel('yield (tons)')
plt.show()
```

### Grouping with Strip Plot

```python
barley.site.value_counts()
```

```python
f.ax=plt.subplots(figsize=(11,5)) # to set size of plot
sns.stripplot(x='site',y='yield',data=barley,jitter=True)
plt.xlabel('Sites', size=15)
plt.ylabel('Yield',size=15)
plt.title('Strip plot',size=17,y=1.03)
plt.suptitle('Site vs Barley yield',size=20,y=1.1)
plt.show()
```

```python
f.ax=plt.subplots(figsize=(15,5))
sns.stripplot(x='variety', y='yield',data=barley, jitter=False)
```

### Swarm Plot

```python
f.ax=plt.subplots(figsize=(12,5))
sns.swarmplot(x='variety',y='yield',data=barley)
```

```python
f.ax=plt.subplots(figsize=(12,5))
sns.swarmplot(x='variety',y='yield',data=barley,hue='site')
```

```python
f.ax=plt.subplots(figsize=(12,5))
sns.swarmplot(x='variety',y='site',data=barley,hue='year',orient='h') #changing the orientation
```

### Box and Violin Plot

```python
#plt.subplot(1,2,1)
sns.boxplot(x='site',y='yield',data=barley)
```

```python
#plt.subplot(1,2,2)
sns.violinplot(x='site',y='yield',data=barley)
```

```python
# placing above 2 plots together
f.ax=plt.subplots(figsize=(12,5))
plt.subplot(1,2,1)
sns.boxplot(x='site',y='yield',data=barley)
plt.subplot(1,2,2)
sns.violinplot(x='site',y='yield',data=barley)
```

```python
f.ax=plt.subplots(figsize=(12,5))
plt.subplot(2,3,1)
sns.boxplot(x='site',y='yield',data=barley)
plt.subplot(2,3,5)
sns.violinplot(x='site',y='yield',data=barley)
```

```python
# combining 2 plots
sns.violinplot(x='site',y='yield',data=barley,inner=None,color='pink')
sns.stripplot(x='site',y='yield',data=barley, size=5,jitter=True)
```

## Bivariate → “two variables” data visualization

### Joint Plot

```python
sns.jointplot(x='year',y='yield',data=barley)
```

### Density Plot

```python
sns.jointplot(x='year',y='yield',data=barley,kind='kde')
```

### Pair Plot

```python
sns.pairplot(barley)
```

```python
sns.pairplot(barley,hue='site')
```
